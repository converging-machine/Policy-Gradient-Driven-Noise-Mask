```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import t
plt.rcParams['font.size'] = 24

# Read the data into a pandas DataFrame
data = pd.read_csv('organamnist.txt', sep=' ', header=None)[[1, 3, 5, 7, 9]]
data.columns = ['F1', 'DATASET', 'INIT', 'MODEL', 'SampleSize']

# Compute the mean and confidence interval for each SampleSize and INIT
results = []
for init, init_data in data.groupby('INIT'):
    for sample_size, sample_size_data in init_data.groupby('SampleSize'):
        f1_scores = sample_size_data['F1']
        mean = f1_scores.mean()
        std_err = f1_scores.std() / np.sqrt(len(f1_scores))
        conf_interval = t.interval(0.95, len(f1_scores) - 1, loc=mean, scale=std_err)
        results.append({'INIT': init, 'SampleSize': sample_size, 'Mean': mean, 'Lower CI': conf_interval[0], 'Upper CI': conf_interval[1]})

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Plot the F1 scores with confidence intervals
plt.figure(figsize=(9, 6))
for init in results_df['INIT'].unique():
    init_data = results_df[results_df['INIT'] == init]
    plt.errorbar(init_data['SampleSize'], init_data['Mean'], yerr=[init_data['Mean'] - init_data['Lower CI'], init_data['Upper CI'] - init_data['Mean']], label=init)
#plt.xlabel('Sample Size')
#plt.ylabel('F1 Score')
plt.title('OrganA - Abdominal CT')
plt.show()
```
